```python
import pandas as pd
```

```python
df = pd.read_parquet('green_tripdata_2025-11.parquet')
```

```python
df.columns
```

```python
pickup_zone_df = pd.read_csv('taxi_zone_lookup.csv')
```

```python
map_dict = dict(zip(*(pickup_zone_df['LocationID'],pickup_zone_df['Zone'])))
```

```python
df['PULocation'] = df['PULocationID'].map(map_dict)
df['DOLocation'] = df['DOLocationID'].map(map_dict)
```

```python
df['lpep_pickup_datetime']
```

```python
df[(df['trip_distance']<=1) & (df['lpep_pickup_datetime'] >= '2025-11-01') & 
    (df['lpep_pickup_datetime'] < '2025-12-01')]
```

```python
df[df['trip_distance']<100].sort_values('trip_distance', ascending=False)
```

```python
df[(df['lpep_pickup_datetime'].astype(str).str.startswith('2025-11-18')) & (df['lpep_dropoff_datetime'].astype(str).str.startswith('2025-11-18'))].groupby('PULocation')['total_amount'].sum().sort_values(ascending=False)
```

```python
df[(df['PULocationID'] == 74) & (df['lpep_pickup_datetime'].astype(str).str.startswith('2025-11'))].groupby('DOLocation')['tip_amount'].max().sort_values(ascending=False)
```
